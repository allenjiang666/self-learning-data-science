# Decision Tree

## Load the data

```python
import pandas as pd 
import numpy as np 
data = pd.read_csv('iris_data.csv')
data.head()
```

## Define the X and y

```python
X = data.drop(['target','label'], axis=1)
y = data['label']
print(X.shape, y.shape)
X.head()
```

## Establish a decision tree model and train

```python
from sklearn import tree
DT = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=5)
DT.fit(X,y)
```

## Evaluate the model

```python
y_predicted = DT.predict(X)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predicted)
accuracy
```

## Visualize the model

```python
data.target.value_counts().
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize=(20,10))
tree.plot_tree(DT, feature_names=X.columns,filled='True', class_names=['setosa','versicolor','virginica'])
plt.show()
```

### try with a smaller min_samples_leaf

```python
DT = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=2)
DT.fit(X,y)
fig2 = plt.figure(figsize=(20,10))
tree.plot_tree(DT, feature_names=X.columns,filled='True', class_names=['setosa','versicolor','virginica'])
plt.show()
```
